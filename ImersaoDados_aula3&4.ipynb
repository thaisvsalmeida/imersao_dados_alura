###Base de dados para ter acesso a latidute e longitude

```python
enderecos = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/ImersaoDados/enderecos.csv")
```

```python
enderecos.head()
```

```python
#Separar apenas SP
enderecos_sp = enderecos.query("sigla_uf == 'SP'")
enderecos_sp
```

```python
enderecos_sp["rua"] = enderecos_sp["tipo_logr"] + " " + enderecos_sp["logr_nome"]
enderecos_sp["rua"] = enderecos_sp["rua"].str.lower().str.strip()
enderecos_sp.head()
```

```python
dadosFiltrados["apenas_rua"] = dadosFiltrados["Rua"].str.extract(r'(^[\w ]+)')
dadosFiltrados["apenas_rua"] = dadosFiltrados["apenas_rua"].str.lower().str.strip()
dadosFiltrados.head()
```

```python
dados_geo = pd.merge(left = dadosFiltrados, right = enderecos_sp[["rua", "cep", "latitude", "longitude"]], how = "left", left_on = "apenas_rua", right_on = "rua").drop_duplicates(subset = dadosFiltrados.columns).query("cep > 0")
dados_geo
```
