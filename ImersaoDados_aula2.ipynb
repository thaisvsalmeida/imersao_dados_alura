####Gráfico: 
####Valor do Anuncio em Milhões de Reais X Quantidade de Imóveis

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize =(16, 8))
ax = sns.histplot(data = dadosFiltrados, x = "ValorAnuncioFloat", kde = True, color="purple", lw = 1)
ax.lines[0].set_color("red")
ax.set_title("Imóveis em São Paulo", size=20)
ax.set(xlabel='Valor do Anuncio em Milhões de Reais', ylabel='Quantidade de Imóveis')
plt.xlim((-50, 10000000))
ticks = ax.get_xticks()*10**-6
ax.set_xticklabels(ticks)
plt.show()
```

####Preço do metro quadrado por bairro

```python
dadosBairros = dadosFiltrados.groupby("Bairro").sum().reset_index()
dadosBairros["R$/m²"] = (dadosBairros["ValorAnuncioFloat"] / dadosBairros["Metragem"])
dadosBairros
```

```python
plt.figure(figsize =(16, 8))
ax = sns.barplot(data = dadosBairros.head(20), y = "Bairro", x = "R$/m²", lw = 1)
ax.set_title("Imóveis em São Paulo", size=20)
ax.set(xlabel='Preço (R$ / m²)', ylabel='Bairros')
plt.show()
```

###Visualização para analisar quartis, mediana e outliers

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(data = dadosBairros, y = "R$/m²", color = "orange")
ax.set_title("Boxplot de Imóveis em São Paulo", size=20)
plt.ylim((-50, 12000))
plt.show()
```

```python
#Top 10 de bairros que aparecem mais vezes no DataFrame
top_bairros = dadosFiltrados["Bairro"].value_counts().head(10).index
top_bairros
```

```python
plt.figure(figsize=(10,8))
preco_medio_imovel = dadosFiltrados.ValorAnuncioFloat / dadosFiltrados.Metragem
ax = sns.boxplot(data = dadosFiltrados.query('Bairro in @top_bairros'), x='Bairro', y=preco_medio_imovel)
ax.set_title("Boxplot de Imóveis por Bairro em São Paulo", size=20)
ax.set(xlabel="Bairros", ylabel='m² por imóvel')
ax.tick_params(axis='x', rotation=45)
plt.show()
```
